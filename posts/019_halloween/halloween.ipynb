# Takeaways and Conclusion
ChatGPT’s response was organized and easy to follow. It split ideas into categories like Pop Culture/TV & Movies, Music/Pop Stars, Internet/Meme Culture, and Simple/Low-Effort. The structure made it easier to process, though it also made the suggestions feel formulaic. Most of the options were popular and recognizable — like Wednesday Addams or Barbie — showing that the AI relies heavily on what’s trending rather than offering anything especially original.

The AI did take practicality into account, often suggesting easy costumes that didn’t require much effort. For instance, the Euphoria-inspired look depended mostly on makeup, which was convenient but also somewhat predictable.

When ChatGPT offered to make a shortlist of five “perfect mix” options, I agreed. Its follow-up included more details on what each costume would need, but the guidance sometimes felt overly specific. Later, I turned down its offer to make a full “cheat sheet” of stores and items, since I didn’t want it to dictate every part of the process. That decision underscored how easy it is to let the AI take over creative control if you’re not careful.
