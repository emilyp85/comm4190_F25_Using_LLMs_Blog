# Analysis
Overall, Gemini's photo editing skills exceeded my expectations. Google Gemini's ability to provide a photo was faster than that of ChatGPT, which I really appreciated; I didn't want to wait minutes every time I requested an edit.

In my first edit, while Gemini successfully achieved the bluish-gray lighting I asked for, the result looked slightly desaturated and black and white_almost monochromatic. this showed me that Gemini, while understanding the "color tone" direction, did not understand the nuance of preserving natural skin tones and small details.

The second edit demonstrated a significant improvement in accuracy. Gemini managed to isolate and preserve the original colors of myself and the floor while still adjusting the background lighting to match my first request. This result looked much more professional and cohesive.

Finally, the third and final edit achieved the exact edit and aesthetic I was looking for. The lowered exposure and subtle increase in blue hues gave the image of a moody, cohesive vibe that would better fit my Instagram profile. The shadows and highlights also appeared balanced, maintaing a realistic feel rather than an AI feel.

In conclusion, Gemini's editing quality was surprisingly high for an AI model when compared to ChatGPT.
