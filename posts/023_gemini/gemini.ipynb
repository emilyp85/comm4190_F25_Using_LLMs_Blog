# Introduction
In my previous blog, I tried to generate a professional headshot using ChatGPT. My takeaway from that blog was that even with a clear reference photo, ChatGPT’s generated headshots produced only a vague resemblance instead of accurately reflecting my facial features. I want to see if that holds true for Google Gemini as well.

In order to keep the experiment the same and control any external factors that can influence the outcome of my experiment, I will be using the same prompt.

Thus, this is the prompt I will upload to Gemini.
> Can you generate a professional headshot for me using this image?

![](headshot.JPG)

Hopefully, the image that Gemini generates for me is better than the one ChatGPT provided. And I mean "better" in the sense that the girl in the photo looks more like me and is more realistic than cartoonish.

# Takeaways
In my opinion, the last one that Gemini gave me was the most realistic and similar looking to me. One thing I noticed was that, unlike ChatGPT, Gemini did not ask me for clarification on what I am looking for. While I still provided one regardless, I thought that maybe Gemini’s approach to image generation relies more on its internal assumptions about what a “professional headshot” should look like, rather than on interactive clarification. ChatGPT, on the other hand, asked follow-up questions to refine its understanding of my request, which made the process feel more collaborative.

That being said, Gemini’s final image did come closer to what I was looking for—it appeared slightly more realistic and less stylized than ChatGPT’s outputs.

I also noticed that Gemini is quicker at generating images than ChatGPT and the "texture" of the images were much more realistic (hence why I liked Gemini's AI-generated photos more than ChatGPT's). Yet, even if I did prefer Gemini's, at the end of the day, you can tell that the photo was AI-generated.

Overall, this experiment reaffirmed my earlier takeaway: while AI can approximate the aesthetics of a professional headshot, it cannot yet replicate the subtleties of a real person’s face.
