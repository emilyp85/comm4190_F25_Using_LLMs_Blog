# Questions and Answers
1) **What does she suggest about the relationship between beauty standards and algorithmic accuracy?**
Bolamwini argues that algorithmic systems, particularly facial recognition technologies, often perform better on faces that align with conventional beauty standards. These standards—shaped historically by Eurocentric ideals—skew training datasets toward Eurocentric features like lighter skin and certain noses. As a result, AI systems are less accurate at recognizing darker-skinned faces, nonbinary or gender-nonconforming individuals, and faces that fall outside these normative standards. She suggests that this shows how technology inherits societal biases rather than being inherently "objective."
   
2) **How does she use the idea of visibility vs. erasure—both in terms of race and credibility—as a recurring theme?**
Buolamwini repeatedly highlights the tension between visibility and erasure. In terms of race, she shows how people of color are often "erased" by AI systems. Similarly, in terms of credibility, marginalized voices and experiences are often ignored or dismissed in conversations like AI ethics and policy. By using the idea of visibility vs. erasure, she emphasizes that invisibility isn't just a societal issue but is technologically mediated, affecting both recognition and representation.

3) **How does she illustrate the dangers of assuming technology as neutral?**
The author illustrates the risks of assuming AI is neutral through multiple examples. She shows that algorithms reflect the data they are trained on and the priorities of their designers. For instance, facial recognition systems may misidentify women of color at disproportionately high rates, not because the technology is flawed in principle, but because the datasets and development processes are biased. By documenting these failures, she warns that treating technology as neutral can hide structural inequalities, reinforce discrimination, and lead to real-world harms.
