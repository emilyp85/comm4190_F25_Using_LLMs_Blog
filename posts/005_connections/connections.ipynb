I play Connections almost every morning. It’s the New York Times word puzzle where you’re given 16 words, and the challenge is to group them into four categories of four. Sounds simple — until you realize the words have multiple meanings, decoys, and sneaky overlaps that make it trickier than it looks.

This time, I decided to test a new teammate: ChatGPT. Could it solve the puzzle faster than I could?

# Input the Command
I wanted to see if ChatGPT can solve the New York Times Connections Game. Thus, I copied the day's words into ChatGPT with a single prompt.

```
Can you solve the NYT connections game? Here are the words:
BANK, COAST, SHORE, STRAND, CHEMISTRY, CONNECTION, FIREWORKS, SPARK, CARPET, LAMINATE, TILE, WOOD, BANSHEE, HEARTBREAKER, PIP, WAILER
```

ChatGPT thought for a few seconds and then gave me this:

```
Thought for 13s

Nice — here's a clean solve with groups and category names:

Groups

Shore / Edge (coastline/banks)
BANK, COAST, SHORE, STRAND

Romantic spark / chemistry (things that signal attraction)
CHEMISTRY, CONNECTION, SPARK, FIREWORKS

Flooring materials
CARPET, LAMINATE, TILE, WOOD

Band-name roots (singular forms of group names used in band names: The ___s)
BANSHEE, HEARTBREAKER, PIP, WAILER

```
# Testing It Out
I typed these into the game. And... it worked! All four groups, correct.

# How It Solved It
When I asked ChatGPT how it reached the answer, it broke its process down step by step:

```
Nice — here's a clear, non-secret walkthrough of how I solved it.

1. **Look for obvious semantic clusters.**
   I scanned the full list for small groups of words that share a single, simple meaning. Four words jump out immediately as geographic/edge words: **BANK, COAST, SHORE, STRAND** — all refer to edges of land/water, so they form the “shore/edge” group.

2. **Spot metaphorical/figurative connections.**
   Some words weren’t literal but are commonly used as metaphors for the same idea. **CHEMISTRY, CONNECTION, SPARK, FIREWORKS** are all used to describe romantic or interpersonal attraction (e.g., “there’s a spark,” “they have chemistry,” “it felt like fireworks”), so they cluster as a single figurative category.

3. **Group by concrete category / function.**
   **CARPET, LAMINATE, TILE, WOOD** are all flooring materials — a tidy, concrete category that’s easy to see once you notice two or three of them.

4. **Recognize cultural/name patterns.**
   The last four required a bit of cultural knowledge: **BANSHEE, HEARTBREAKER, PIP, WAILER** all appear in band names (or groups): e.g., *Siouxsie and the Banshees*, *Tom Petty and the Heartbreakers*, *Gladys Knight & The Pips*, *Bob Marley & The Wailers*. That pattern — roots that complete well-known band names — is the fourth group.

5. **Check for overlap or alternate fits.**
   After grouping, I scanned for words that might belong to more than one group and chose the group where the meaning is strongest and clearest. Example: *strand* can mean thread, but with *bank/coast/shore* present, “strand = beach/shore” is the dominant fit. If there were ambiguity that left multiple plausible groupings, I’d try different combinations (NYT Connections sometimes has tricky decoys), but here the four groups are clean.

That’s it — a mix of looking for tight literal sets, spotting figurative uses, and recognizing cultural naming conventions.

```
That’s not far off from how a human would explain their own reasoning — but instead of trial and error on the board and using up limited attempts, the AI laid out categories almost instantly.

# AI vs. Human Play Style

What struck me is how differently AI and humans approach the same puzzle:

* Humans often guess, regroup, and backtrack when categories overlap.
* ChatGPT skipped the guessing and presented neat, confident sets, backed with cultural references and metaphors.

It didn’t feel like it was “playing” so much as sorting words into buckets based on context clues it had seen before.

ChatGPT’s categories were perfect this time, but what impressed me was its confidence. I am very hesitant to submmit my answers in this game but ChatGPT was not.

The cultural (the hardest!) category was the most surprising. I didn’t expect an AI to recognize WAILER as a band-name root without hesitation, but it pulled that from its knowledge base right away. Even I didn't know that those group of words could form a category.

# Should You Play with an AI Teammate?
Would I use ChatGPT every day to “cheat” at Connections? Probably not — that would ruin the fun. But it was fascinating to see how it *thinks* through puzzles like this.

It also raises a bigger question: was this “reasoning” or just sophisticated pattern-matching? To me, it felt like a blend. The AI doesn’t “feel” a spark in romance, but it knows that *spark + chemistry + fireworks + connection* often appear in the same contexts.

This made me wonder — how well could ChatGPT handle other NYT puzzles? Wordle seems too easy, but what about Spelling Bee? Or even the Mini Crossword? Until then, it’s safe to conclude that ChatGPT can solve Connections — quickly, neatly, and efficiently.
